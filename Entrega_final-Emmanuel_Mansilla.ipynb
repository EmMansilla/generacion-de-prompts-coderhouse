## 1. Problema a abordar

El proyecto está orientado a empresas pequeñas, mayormente de alcance
regional. En ellas, es imperioso automatizar actividades y modernizar su gestión,
utilizando metodologías que permitan completar proyectos de calidad sin
excederse en el tiempo y esto se logrará con un plan de trabajo organizado y
controlado. A su vez, estas empresas suelen disponer de menos recursos
económicos que puedan destinar a capacitación de empleados en habilidades
técnicas que no sean las operativas propias de la actividad.

## 2. Propuesta de solución

A partir del problema enunciado anteriormente, contar con el soporte de un
asistente que utiliza inteligencia artificial permitirá obtener sugerencias rápidas
referidas a la planificación para utilizar en nuevos proyectos. La ventaja del
modelo para estas instituciones radica principalmente en:  
● Su base de conocimiento en tareas de planificación de proyectos podría
ser mayor que la de los empleados.  
● Disponer de sugerencias de planificación optimizará los tiempos
administrativos, lo que resulta favorable para la rentabilidad de la
Organización.

## 5. Implementación

```python
import openai
```

```python
openai.api_key = ""
```

```python
contexto = "Eres experto en metodología ágiles, especializado en Scrum. Tus respuestas serán breves y precisas."
```

```python
continue_loop = True

# Bienvenida
print("Hola! Soy un asistente virtual de planificación de proyectos. Por favor, ingresa los siguientes datos para obtener el cronograma de actividades propuestos.")

#Solicitud del nombre de proyecto.
nombre_proyecto = input("Ingresa el nombre del proyecto: ")
print("Muchas gracias!")

#Solicitud del objetivo del proyecto.
objetivo_proyecto = input("Ahora describe brevemente cuál es el objetivo del proyecto, por favor:  ")
print("Comprendo.")

#Solicitud de Fecha de inicio.
fecha_inicio_proyecto = input("Ingresa la fecha de inicio del proyecto en formato dd/mm/yyyy:  ")
print("Perfecto! Ya estamos por terminar.")

#Solicitud de duración del proyecto.
duracion_proyecto = input("Indica la cantidad de semanas que tienes como plazo hasta la entrega:  ")
print("Comprendido! Solo queda un paso más.")

#Solicitud de cantidad de participantes.
participantes_proyecto = input("¿Cuántas personas participarán en el proyecto?:  ")
print("Gracias nuevamente! En instantes tendrás disponible la información.")


#prompt (text/user's question)
prompt = f"Diseñar la planificación de un proyecto mostrando únicamente la lista de participantes con su profesión y un cronograma en formato de tabla donde se detalle las tareas a realizar con sus fechas de inicio y fin y la profesión de la persona responsable. Usa nombres genéricos como Persona_1 o Persona_2. El nombre del proyecto es {nombre_proyecto}. El objetivo es {objetivo_proyecto}. La fecha de inicio del proyecto es {fecha_inicio_proyecto}. La duración del proyecto en cantidad de semanas es {duracion_proyecto}. El equipo de trabajo se conforma por {participantes_proyecto} personas."

#Consulta
conversation = [ {"role": "system", "content": contexto}, {"role": "user", "content": prompt}]

response = openai.ChatCompletion.create(model='gpt-3.5-turbo',messages=conversation,max_tokens=1000)

#Storage the response in a variable to use it
phase = response['choices'][0]['message']['content'].strip()
    
#Use the variable
print(phase)
print("¡IMPORTANTE!: Recuerda que el resultado es una guía que puedes modificar según las necesidades de la empresa. ¡Éxitos en la implementación!")
```

## 6. Resultados

La respuesta obtenida da como resultado una guía esperada de las actividades principales, junto con los plazos para cada tarea y el perfil de la persona necesaria para llevarlo a cabo. En el ejemplo, el proyecto general podría luego particionarse de manera tal que cada tarea sea procesada como un proyecto diferente y obteniendo así más detalle para cada etapa.
